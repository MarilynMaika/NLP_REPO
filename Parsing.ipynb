###Syntactic parsing:
is the process of analyzing the structure of a sentence to determine how its components (words) relate to each other grammatically.<br>
### Key Concepts:
Parse Trees: Visual representation of sentence structure.
<ul><li>Constituency Parsing: Decomposes a sentence into subphrases (e.g., noun phrases, verb phrases).Focuses on breaking the sentence into constituent parts or phrases.
<ul><li>Parse Tree: Shows how sentences are hierarchically organized into phrases.
<li>Phrase Structure Rules: Explain how grammar rules (like S -> NP VP) govern sentence structure.</ul>
<li>Dependency Parsing: Focuses on relationships between words (e.g., "subject" or "object").

## Constituency Parsing

```python
import nltk
from nltk import CFG

# Define a simple grammar using CFG (Context-Free Grammar)
grammar = CFG.fromstring("""
    S -> NP VP
    NP -> Det N | Det N PP
    VP -> V NP | V NP PP
    PP -> P NP
    Det -> 'the' | 'a'
    N -> 'cat' | 'mouse' | 'garden'| 'police'|
    V -> 'chased' | 'saw'| 'teargassed'
    P -> 'in'
""")
# Create a parser based on the grammar
parser = nltk.ChartParser(grammar)

# Parse a sentence
sentence = "the cat chased the mouse".split()
for tree in parser.parse(sentence):
    print(tree)
    #tree.draw()  # This will visualize the parse tree
    tree.pretty_print() # Print the tree to console
```

## Dependency Parsing

```python
#Introduce SpaCy for dependency parsing, which is more advanced and widely used in modern NLP applications.
import spacy

# Load the English model in SpaCy
nlp = spacy.load('en_core_web_sm')

# Parse a sentence
doc = nlp("The cat swiftly climbed the tall tree to catch three elusive birds")

# Print dependency relations
for token in doc:
    print(f'{token.text} -> {token.head.text} ({token.dep_})')

# Visualize the dependency tree (requires the 'spacy.displacy' module)
spacy.displacy.render(doc, style="dep")
```

```python
# Import required libraries
import nltk
nltk.download('punkt_tab')

# Download the English version of the averaged perceptron tagger
nltk.download('averaged_perceptron_tagger_eng')
from nltk import pos_tag, word_tokenize, RegexpParser

# Example text
sample_text = "The quick brown fox jumps over the lazy dog"

# Find all parts of speech in above sentence
tagged = pos_tag(word_tokenize(sample_text))

#Extract all parts of speech from any text
chunker = RegexpParser("""
                    NP: {<DT>?<JJ>*<NN>} #To extract Noun Phrases
                    P: {<IN>}             #To extract Prepositions
                    V: {<V.*>}            #To extract Verbs
                    PP: {<P> <NP>}       #To extract Prepositional Phrases  #Removed newline and <p> tag
                    VP: {<V> <NP|PP>*}    #To extract Verb Phrases
                    """)

# Print all parts of speech in above sentence
output = chunker.parse(tagged)
print("After Extracting\n", output)
```
