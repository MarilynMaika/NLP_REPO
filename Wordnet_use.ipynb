```python
from math import hypot
import nltk
from nltk.corpus import wordnet as wn

#download wordnet data (only needed once)
nltk.download("wordnet")


# 1 look up a word to get it's synsets

word="bank"

synsets=wn.synsets(word)

print(f"Total synsets found for '{word}': {len(synsets)}\n")
#print(synsets)

print("---------------------------------------------------------")


#2. inspect the primary noun synset for 'word of interest'

primary_synset=synsets[1]

print(f"\nPrimary synset ID: {primary_synset.name()}")
print(f"Primary synset definition: {primary_synset.definition()}")
print(f"Primary synset examples: {primary_synset.examples()}")


print("---------------------------------------------------------")



#3. Extract all synonyms belonging to this synset
synonyms=[lemma.name() for lemma in primary_synset.lemmas()]

print(f"\nSynonyms for '{primary_synset.name()}': {synonyms}")





print("---------------------------------------------------------")

# 4. Find  broader concepts (Hypernyms)

hypernyms=primary_synset.hypernyms()

if hypernyms:
  print(f'Broader concept (Hypernym): {hypernyms[0].name()}')
  print(f'Hypernym definition: {hypernyms[0].definition()}')
```
