# Lab 1: Word Sense Disambiguation (WSD)
Word Sense Disambiguation (WSD) is the process of determining which meaning of a word is activated
by its use in a particular context. For example, in the sentence "I went to the bank," WSD helps
determine whether "bank" refers to a financial institution or the side of a river.<br>
<b>Approaches to WSD:</b>
<ol>
<li>. Knowledge-based Methods: Use predefined resources like WordNet to disambiguate words.
- Lesk Algorithm: Disambiguates word senses by comparing overlaps between the context of the word
and the definitions of the possible senses in a dictionary or WordNet.
<li>. Supervised Methods: Use labeled data to train a machine learning model for disambiguation.
- Naive Bayes: A popular classifier that can be used for WSD by modeling the likelihood of word senses
based on their context.
<li>. Unsupervised Methods: These methods cluster words into different senses without labeled data.
- Contextual Embeddings: Modern NLP techniques (e.g., BERT) can infer word senses from context
using contextual embeddings, significantly improving WSD performance.

<b>Learning objectives:</b>
<ul><li>Understand what a synset is and how to inspect its definition.
<li>Apply the Lesk algorithm to disambiguate a target word.
<li>Compare the chosen sense against all available senses.

```python
import nltk
nltk.download('punkt_tab')
nltk.download('wordnet')
from nltk.wsd import lesk
from nltk.corpus import wordnet as wn
```

```python
#Explore synsets for an ambiguous word
word = "bank"
synsets = wn.synsets(word)
print(f"\nAll synsets for '{word}':")
for i, syn in enumerate(synsets, 1):
  print(f"  [{i}] {syn.name():<25} POS: {syn.pos()} | "
              f"Def: {syn.definition()}")
```

## 1. Lesk Algorithm for WSD
- In Python, we can implement a basic version of the Lesk Algorithm using WordNet:

```python
sentence = "I went to the bank to deposit money."

word_sense = lesk(sentence.split(), 'bank')

print(f"Sense: {word_sense}")
print(f"Definition: {word_sense.definition()}")
```

```python
# Apply Lesk to different sentences
from nltk.tokenize import word_tokenize
sentences = [
        ("I deposited money at the bank.", "bank"),
        ("She sat on the bank of the river.", "bank"),
        ("The bass guitar has a deep sound.", "bass"),
        ("He caught a large bass in the lake.", "bass"),
        ("The crane lifted the steel beam.", "crane"),
        ("A crane flew over the wetlands.", "crane"),
    ]
for sentence, target in sentences:
    tokens = word_tokenize(sentence.lower())
    best_sense = lesk(tokens, target)
    if best_sense:
          print(f"\n  Sentence : {sentence}")
          print(f"  Target   : '{target}'")
          print(f"  Synset   : {best_sense.name()}")
          print(f"  Meaning  : {best_sense.definition()}")
    else:
          print(f"\n  No synset found for '{target}' in: {sentence}")
```

## 2. Semantic Analysis Using WordNet
WordNet is an invaluable resource for analyzing the semantic relationships between words. It groups
words into synsets, which are sets of synonyms sharing the same meaning, and defines various
relationships such as hyponymy and hypernymy.

# Lab 2: SEMANTIC SIMILARITY
Background:<br>
WordNet's hierarchical structure lets us measure how closely related two concepts are. Three popular metrics are:<br>
<ul><li>Path similarity   — inverse of the shortest path length between synsets in the taxonomy.
<li>Wu-Palmer (WUP)  — considers the depth of the synsets and their
least common hypernym (LCH).
<li>Leacock-Chodorow — uses path length + taxonomy depth.</ul>

<b>Learning objectives:</b>
<ul><li> Compute pairwise semantic similarity using multiple metrics.
<li>Understand when words are considered semantically close or distant.
<li>Build a simple similarity comparison table. </ul>

##PART A — Similarity between two words (first synsets)

```python
word_pairs = [
        ("dog",   "cat"),
        ("dog",   "wolf"),
        ("dog",   "automobile"),
        ("car",   "vehicle"),
        ("happy", "joyful"),
        ("happy", "sad"),
        ("king",  "queen"),
        ("bank",  "river"),
    ]

header = f"{'Word 1':<12} {'Word 2':<12} {'Path':>8} {'WUP':>8} {'LCH':>8}"
print(f"\n  {header}")
print("  " + "-" * len(header))

for w1, w2 in word_pairs:
    syns1 = wn.synsets(w1, pos=wn.NOUN)
    syns2 = wn.synsets(w2, pos=wn.NOUN)
    if not syns1 or not syns2:
       continue
    s1, s2 = syns1[0], syns2[0]
    path = s1.path_similarity(s2)
    wup  = s1.wup_similarity(s2)
    lch  = s1.lch_similarity(s2)
    path_str = f"{path:.4f}" if path  is not None else "  N/A"
    wup_str  = f"{wup:.4f}"  if wup   is not None else "  N/A"
    lch_str  = f"{lch:.4f}"  if lch   is not None else "  N/A"
    print(f"  {w1:<12} {w2:<12} {path_str:>8} {wup_str:>8} {lch_str:>8}")
```

## PART B — Finding the most similar word in a list

```python
def most_similar(target_word, candidates):
        """Return the candidate most similar to target_word (path similarity)."""
        t_syns = wn.synsets(target_word, pos=wn.NOUN)
        if not t_syns:
            return None, 0.0
        t_syn = t_syns[0]
        best_word, best_score = None, -1.0
        for word in candidates:
            c_syns = wn.synsets(word, pos=wn.NOUN)
            if not c_syns:
                continue
            score = t_syn.path_similarity(c_syns[0]) or 0.0
            if score > best_score:
                best_score, best_word = score, word
        return best_word, best_score

queries = [
        ("doctor",    ["nurse", "engineer", "pilot", "teacher"]),
        ("laptop",    ["keyboard", "ocean", "mountain", "thunder"]),
        ("football",  ["tennis", "basketball", "guitar", "cloud"]),
    ]

for target, candidates in queries:
        best, score = most_similar(target, candidates)
        print(f"\n  Target   : '{target}'")
        print(f"  Pool     : {candidates}")
        print(f"  Closest  : '{best}'  (path similarity = {score:.4f})")
```

## PART C — Hypernym path (semantic hierarchy)

```python
def hypernym_chain(word):
        syns = wn.synsets(word, pos=wn.NOUN)
        if not syns:
            return []
        chain = []
        current = syns[0]
        while current.hypernyms():
            current = current.hypernyms()[0]
            chain.append(current.name())
        return chain

for w in ["cat", "car", "laptop"]:
        chain = hypernym_chain(w)
        print(f"\n  '{w}' → {' → '.join(chain)}")
```
